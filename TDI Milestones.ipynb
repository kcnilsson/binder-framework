What is the total number of jobs created for businesses in Queens?

```python
#Total number of jobs created for businesses in Queens
queens_jobs = df.set_index('Borough').loc['Queens']['Job created'].sum()
queens_jobs
```

How many different unique email domains names are there in the data set?

```python
df['Domain'] = df['company email'].dropna().apply(lambda x:x.partition("@")[2])
unique_domain = df['Domain'].nunique()
unique_domain
```

Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

```python
average_savings_and_total_jobs = (df.groupby('NTA')[['Company Name', 'Job created', 'Total Savings', 'NTA']]
 .filter(lambda x:x['Company Name'].count()>5).groupby(['NTA']).agg({'Total Savings':'mean', 'Job created':'sum'})
 .rename(columns={'Total Savings':'Average Savings', 'Job created':'Total Jobs Created'}))
average_savings_and_total_jobs
```

```python
average_savings_and_total_jobs.to_csv('average_savings_total_jobs.csv')
```

```python
x = average_savings_and_total_jobs['Total Jobs Created']
y = average_savings_and_total_jobs['Average Savings']
```

```python
import seaborn as sns
```

```python
sns.set_theme(style="white")

# Plot miles per gallon against horsepower with other semantics
sns.relplot(x='Total Jobs Created', y='Average Savings', hue="NTA",
            height=6, data=average_savings_and_total_jobs)
plt.title('Standard')
sns.relplot(x='Total Jobs Created', y=np.log10(y), hue="NTA",data=average_savings_and_total_jobs)
plt.title('Logarithmic');
```

```python
sns.histplot(data=average_savings_and_total_jobs, x=np.log10(y))
```

```python
df['Month'] = pd.DatetimeIndex(df['Effective Date']).month
```

```python
total_jobs_created_by_month = (df.groupby('NTA')[['Company Name', 'Job created', 'Total Savings', 'Month', 'NTA']]
 .filter(lambda x:x['Company Name'].count()>5).groupby(['NTA', 'Month']).agg({'Job created':'sum'})
 .rename(columns={'Job created':'Total Jobs Created'}))
```
